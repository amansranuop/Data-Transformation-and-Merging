By summing the values of individual answers (answer value), Gad score gets caculated for each group on a specific date.

```python
# Group by 'group_identifier' and 'Date', summing the 'answer_value' to get the GAD score
gad_scores = df.groupby(['group_identifier', 'Date'])['answer_value'].sum().reset_index()
```

```python
# Rename columns to match the format in postprocessedGad.csv
gad_scores.rename(columns={'answer_value': 'General Anxiety Disorder Score'}, inplace=True)
```

```python
gad_scores['General Anxiety Disorder Score'].describe()
```

### Change scores into a new csv file

```python
# Saving the processed data to a CSV file
gad_scores.to_csv('processedGad.csv', index=False)
```

## Part 2:

```python
garmin = pd.read_csv("garmin.csv")
daily = pd.read_csv("postprocessedDaily.csv")
```

```python
garmin.head(5)
```

```python
daily.head(5)
```

```python
data.head(5)
```

```python
# Merge garmin.csv with postprocessedDaily.csv
merged_garmin_daily = pd.merge(garmin, daily, on=['group_identifier', 'Date'], how='outer')

# Sort the merged data by 'group_identifier' and 'Date'
merged_garmin_daily.sort_values(by=['group_identifier', 'Date'], inplace=True)

# Display the first few rows to ensure it's merged correctly
print(merged_garmin_daily.head(5))
```

```python
# Merge the GAD scores with the already merged Garmin + Daily data
# Perform the merge using an outer join to retain all rows
merged_with_gad = pd.merge(merged_garmin_daily, data, on=['group_identifier', 'Date'], how='outer')

# Sort the data by 'group_identifier' and 'Date' for forward-filling
merged_with_gad.sort_values(by=['group_identifier', 'Date'], inplace=True)

# Forward-fill the GAD scores for up to 30 days
merged_with_gad['General Anxiety Disorder Score'] = merged_with_gad.groupby('group_identifier')['General Anxiety Disorder Score'].ffill(limit=30)

# Display the first few rows to ensure it’s working as expected
print(merged_with_gad.head())
```

```python
merged_with_gad.to_csv('final_merged_data.csv', index=False)
```
